When selecting datasets to analyze the effect of weather on retail sales, there are several issues which must be considered:

1. **Selecting the <u>time granularity</u> of the sales and weather data**</br>
- The first question to ask is whether it is better to select data about weather and sales which have data points representing aggregate values daily, weekly, monthly, quarterly or annually. Determining the time granularity of the data depends to a large extent on the research questions we are trying to answer and the kinds of data available. Different granularities come with different advantages and disadvantages which will be touched on briefly here.<br>
- The advantages of using smaller granularities such as weekly or even daily data are that this enables us to analyze the impact of shorter-term weather fluctuations on sales volumes. It allows capturing rapid shifts in consumer behaviour and volatiliy in demand [(*CashFlowInventory.com Blog*)](https://cashflowinventory.com/blog/monthly-vs-weekly-inventory-forecasting/). For instance, it is possible to account for changes in consumer behaviour during a few exceptionally rainy or stormy days in the middle of summer.
- Nonetheless, there are important disadvantages of using weekly data over monthly or quarterly data that must be taken into account. The most pertinent issue is that the Office of National Statistics and other sources of retail sales data do not usually provide monthly values for long time periods (over decades), and thus it is difficult to find weekly data. In this study, the focus of the research questions is largely on how increasingly hot summers (due to global warming) might have impacted sales figures, therefore it is important that the time periods spans at least a couple of decades. Furthermore, weekly values can lead to a large volume of data which significantly increases the time taken to train and validate the models. Another advantage of using monthly over weekly data is that these figures smooth out some of the noise in the data, thus making it easier to identify patterns due to decreased volatility [(*CashFlowInventory.com Blog*)](https://cashflowinventory.com/blog/monthly-vs-weekly-inventory-forecasting/).
- Using annual data might not be the optimal granularity for this project, as this would mean that the data would have a significant lack of detail about how sales patterns change throughout the year based on the seasonal temperatures.
<br></br> Bearing this in mind, in this project monthly data will be used for the initial research to try to identify key trends and patterns. However, if an overall relationship can be established as a foundation, it might be a good idea to extend the research to weekly data in the future, provided that public and open datasets for these sales figures can be found.

2. **Seasonally or Non-Seasonally Adjusted Sales Figures?**
The ONS offers retail sales figures that have been seasonally-adjusted to account for events such as Christmas or Bank Holidays, where retail sales are expected to be affected by such events. When selecting which dataset to use, it first had to be decided whether to use seasonally-adjusted or non-seasonally-adjusted data. Seasonally-adjusted data was chosen as preferable in this context to non-adjusted data for the reason that including increased sales figures before Christmas or Easter would introduce another complicating factor which would make it more difficult to discern the relationship between weather and sales figures. This is how the seasonal-adjustment process is defined on the ONS website:
> "Seasonally adjusted estimates are derived by estimating and removing calendar effects (for example, Easter moving between March and April) and seasonal effects (for example, increased spending in December as a result of Christmas) from the non-seasonally adjusted estimates. Seasonal adjustment is performed each month and reviewed each year using the standard, widely used software, X-13-ARIMA." </br>
> [Office for National Statistics](https://www.ons.gov.uk/businessindustryandtrade/retailindustry/methodologies/retailsalesindexrsiqmi)


3. **Selecting the Time Range: 1988 to 2023**</br>
The ONS's datasets on retail sales span this period of time, which is a relatively long period to analyze this relationship over. The advantage of using a long time period here is that it makes it possible to discern how inflation-adjusted retail sales have been affected by the increasing temperatures of the last few decades. For instance, focusing only on the past 5 or 10 years with weekly-granularity would make it difficult to identify the effect of this. A longer time period minimizes the impact of short-term fluctuations and can hopefully pick up on more general trends.

4. **Table Selection from the ONS RSI Dataset: Volume or Value?**
- The [ONS Retail Sales Index](https://www.ons.gov.uk/businessindustryandtrade/retailindustry/datasets/retailsalesindexreferencetables) dataset will be used here. This is a complex spreadsheet with many tables and pages, displaying different metrics and units for measuring changes in volume and value of sales. Table 1M shows the "chained **volume** of retail sales (seasonally adjusted)" as an index with the base year set to 2019 for which the index is 100. As such, a value of 110 for a certain month means that the sales for that month were 10% higher than the average sales in the base year, and 70 means that the sales were 30% lower than the average for the base year.
- The CPSA Retail Sales Index tables contain the retail sales **value** (taking into account prices as well as amount of sales) as a percentage change for that month on either the same month a year earlier, 3 months on the same period a year earlier, percentage change on the previous month, and percentage change 3 months on the previous 3 months.
</br>
- The KPSA tables show the sales **volume** or amount using the Index2019=100 index, as well as percentage changes similar to the CPSA tables described above but for volume rather than value.
- The first question here is whether the data used in this analysis should account for the volume or the value of monthly retail sales. As the focus here is the economic impact on businesses and how they can maximize profits, value will be chosen rather than volume, especially as the ONS makes it clear that these prices have been adjusted to **account for inflation** across the different sectors:
> "To remove the direct effects of prices changes, value data are deflated to produce RSI volume measures. A separate price deflator is used for each RSI industry at four-digit SIC."
> [ONS: Deflation](https://www.ons.gov.uk/businessindustryandtrade/retailindustry/methodologies/retailsalesindexrsiqmi)

- The second question is whether to use the **2019:100** index for the total sales value or to use **percentage changes** instead. The index is better-suited for exploring long-term patterns, while percentage change may be better suited for exploring short-term dynamics. It is difficult to decide at this stage prior to any analysis which metric would be more useful, therefore in the final dataset we will construct for the regression, it might be helpful to start with the index but to later look at the percentage change on the same month a year earlier and then the percentage change on the previous month to examine the data at a greater level of detail after getting a grasp on longer-term trends first.
