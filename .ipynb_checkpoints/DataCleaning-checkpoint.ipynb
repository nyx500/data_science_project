The retail sales volume tables from the ONS Retail Sales Index spreadsheets which we want to use are KPSA1 and KPSA3
Each of these Excel sheets contains two tables stacked vertically on top of each other.
We want to import the first table (top table) from each of these sheets into a pandas DataFrame as the first step

Docs: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html

How to open Excel files in pandas: https://blog.devgenius.io/reading-excel-files-with-pandas-the-basics-6a6be9cc8763

```python
# This function reads in the correct Excel sheet (KPSA1 for seasonally-adjusted sales volume %-change on same month a year earlier, 
# KPSA3 for %-change on previous month), removes the multi-index including two lines for Agg/SIC and Dataset Identifer codes leaving only meaningful
# column names, and pickles the resulting file for quicker access to DataFrame
def processRetailSalesIndexExcelSheet(fpath, sheetName, nrows, picklePath):
    # Read in the excel file for RSI sales volume
    salesvol_df = pd.read_excel(
        fpath, # filepath
        engine='openpyxl', # support newer Excel file formats.
        sheet_name=sheetName, # sheet name from excel file: will be KPSA1 and KPSA3
        skiprows=5, # Skip first 5 rows because they are not the actual table, but descriptions of the table.
        header = [0, 1, 2], # Header contains 3 lines in the Excel file: column names, then AGG/SIC, then Dataset Identifier Code
        nrows = nrows, # Number of rows in the target table (Excel sheet contains two vertically-stacked tables)
        parse_dates=[0], # Parse the first column of dates: format is'2023 Jun', '2023 May' etc.
        date_format='%Y %B'
    )
    # Drop the multi-index consisting of AGG/SIC and Dataset identifier codes, and leave only the column names in the header
    salesvol_df.columns = salesvol_df.columns.droplevel([1, 2])
    # Save the table as a pickle file for easy loading
    salesvol_df.to_pickle(picklePath)
    # Return the dataframe
    return salesvol_df
```

```python
# Import table KSA1 - Retail Sales Index: Volume Seasonally Adjusted, Percentage Change on Previous Month (for short-term sales fluctuation analysis)
# The actual table is on Excel rows 6 to 438 (430 rows to read in)
ksa1_df = processRetailSalesIndexExcelSheet('./data/retail_sales.xlsx', 'KPSA1', 430, './data/ksa1.pkl')
```

```python
# Now we have to deal with the weather data and import it in as a file, restructure the rows and columns so that 
# each row represents a month like in the sales data (right now each row represents a year and each column a month)

def importWeatherTable(fpath, rows_to_skip, picklePath):
    weather_df = pd.read_csv(
        fpath,
        sep=' ', # even though saved as .csv, these files are actually tsv files and tab-delimited
        skipinitialspace=True, # spacing between values is inconsistent, so set this to true to skip spaces
        index_col=0, # set year-column as index
        skiprows=rows_to_skip, # skip description at the top
        header=0 # set month/season names as columns
    )
    weather_df.to_pickle(picklePath)
    return weather_df
```

```python
airfrost_df = importWeatherTable('./data/air_frost.csv', 5, './data/air_frost.pkl')
heavyraindays_df = importWeatherTable('./data/heavy_raindays.csv', 5, './data/heavy_raindays.pkl')
maxtemp_df = importWeatherTable('./data/max_temp.csv', 5, './data/max_temp.pkl')
mintemp_df = importWeatherTable('./data/min_temp.csv', 5, './data/min_temp.pkl')
meantemp_df = importWeatherTable('./data/mean_temp.csv', 5, './data/mean_temp.pkl')
rainfall_df = importWeatherTable('./data/rainfall.csv', 5, './data/rainfall.pkl')
sunshine_df = importWeatherTable('./data/sunshine.csv', 5, './data/sunshine.pkl')
sunshine_df
```

```python
# Function which drops cols following 'dec' in each weather table, getting rid of seasons/annual attributes and keeping only months
def dropSeasons(df):
    df2 = df.iloc[:, 0:12]
    return df2

airfrost_df2 = dropSeasons(airfrost_df)
heavyraindays_df2 = dropSeasons(heavyraindays_df)
maxtemp_df2 = dropSeasons(maxtemp_df)
mintemp_df2 = dropSeasons(mintemp_df)
meantemp_df2 = dropSeasons(meantemp_df)
rainfall_df2 = dropSeasons(rainfall_df)
sunshine_df2 = dropSeasons(sunshine_df)

# Check for missing values before calculating percentage changes
weather_dataframes = [airfrost_df2, heavyraindays_df2, maxtemp_df2, mintemp_df2, meantemp_df2, rainfall_df2, sunshine_df2]
for df in weather_dataframes:
    nullcounts = df.isnull().sum()
    for nullcount in nullcounts:
        if nullcount != 0:
            print(nullcount)
```

```python
# Melts the DataFrame from wide to long format
# Attribution: https://www.kdnuggets.com/2023/03/beginner-guide-pandas-melt-function.html
def meltWeatherFrame(df, parameter_name):
    # Lists all the years and then months inorder, so all the entries for Januaries in the column, then Februaries etc.
    melted_df = pd.melt(df, id_vars=['year'], var_name='month', value_name=parameter_name.title())
    print(f"{parameter_name.title()}: Melted Dataframe")
    print(melted_df.head(10))
    return melted_df
```

```python
# Create a new column in the dataframe which concatenates the year and the month
def createTimePeriodColumn(df):
    df["Time Period"] = df["year"].astype(str) + " " + df["month"]
    print("Time Period column added:")
    print(df.head(10))
    return df
```

```python
# Convert the string of form '1960 jan' to 1960-01-01 to enable sorting and percentage-change calculations later
def convertTimePeriodColumnToDateTime(df):
    # Convert "1960 jan" string to 1960-01-01 to make possible sorting by month
    df['Time Period'] = pd.to_datetime(df['Time Period'],  format='%Y %b')
    sorted_df = df.sort_values(by='Time Period')
    print("Sorted DF:")
    print(sorted_df.head(10))
    return sorted_df
```

```python
# Currently, the weather tables are in wide-format instead of long-format
# Re-structure them using the following function so that the format matches that of the sales volume tables
def wideToLongWeatherData(df, parameter_name):
    # Make the index column into a normal column
    df.reset_index(inplace=True)
    melted_df = meltWeatherFrame(df, parameter_name)
    melted_df_with_timeperiod = createTimePeriodColumn(melted_df)
    sorted_df = convertTimePeriodColumnToDateTime(melted_df_with_timeperiod)
    sorted_df_with_date_index = sorted_df.set_index('Time Period')
    sorted_df_with_date_index.drop(['year', 'month'], axis=1, inplace=True)
    print("Dataframe now with date year-month column:")
    print(sorted_df_with_date_index.head(10))
    return sorted_df_with_date_index
```

```python
airfrost_long = wideToLongWeatherData(airfrost_df2, "days of airfrost")
heavyraindays_long = wideToLongWeatherData(heavyraindays_df2, "days with over 1mm rain")
maxtemp_long = wideToLongWeatherData(maxtemp_df2, "max temperature (C)")
mintemp_long = wideToLongWeatherData(mintemp_df2, "min temperature (C)")
meantemp_long = wideToLongWeatherData(meantemp_df2, "mean temperature (C)")
rainfall_long = wideToLongWeatherData(rainfall_df2, "rainfall in mm ")
sunshine_long = wideToLongWeatherData(sunshine_df2, "hours of sunshine")
```

```python
# We now need to select the columns which go from January 1988 to be consistent with the sales data
def selectRowsFrom1988(df):
    df_shortened = df.loc['1988-01-01':]
    return df_shortened

```

```python
# Now we can use the strf function to conver the dates from the yyyy-mm-dd formats into the yyyy-monthname format used in the sales dataframe
def convertDates(df):
    df.index = df.index.strftime('%Y %b')

dataframes_to_convert = [
    airfrost_long_1988,
    heavyraindays_long_1988,
    mintemp_long_long_1988, 
    maxtemp_long_1988,
    meantemp_long_1988, 
    rainfall_long_1988,
    sunshine_long_1988
]
```

```python
all_combined = pd.concat([combined_sales_df, weather_combined], axis=1)
all_combined.head(20)
```

```python
import matplotlib.pyplot as plt
```

```python
# Now we can see that the first few rows of the KPSA1 sales data and the last two months' rows for the KPSA3 
# sales data have some NaN missing values, which are not well-suited to machine-learning algorithms...
# The problem here is that if we use the pandas bfill or ffill functions, this will not reflect possible sales-volume change from month-to-month and
# skew the results of the model. Using the mean is possible but will not really capture the pattern of fluctuations between months either. Pandas offers various
# ways to 'interpolate' these missing values such as 'linear' interpolation, 'time' interpolation for time-series
# We will compare how interpolating with different parameters or filling with the column mean could change the results of our regression algorithm!

all_combined_mean = all_combined.fillna(all_combined.mean(axis=0))
all_combined_mean.head(10) # All the first months for KPSA1 are now the same, which means it is not really useful for looking at monthly percentage change
plt.plot(all_combined_mean.iloc[:, 0])
```

```python
# Attribution: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html
all_combined_linear_interpolated = all_combined.interpolate(method='linear', limit_direction='both', axis=0)
all_combined_linear_interpolated
plt.plot(all_combined_linear_interpolated.iloc[:, 0])
```

```python
# Interpolate a time series here
all_combined_time = all_combined.copy()
all_combined_time.index = all_combined_time.index = pd.to_datetime(all_combined_time.index)
all_combined_time_interpolated = all_combined_time.interpolate(method='time', limit_direction='both', axis=0)
plt.plot(all_combined_time_interpolated.iloc[:, 0])
```

```python
all_combined_time_spline_interpolated = all_combined_time.interpolate(method='spline', order =1, limit_direction='both', axis=0)
all_combined_time_spline_interpolated 
plt.plot(all_combined_time_spline_interpolated .iloc[:, 0])
```

```python
all_combined_time_nearest_interpolated = all_combined_time.interpolate(method='nearest', limit_direction='both', axis=0)
all_combined_time_nearest_interpolated
plt.plot(all_combined_time_nearest_interpolated.iloc[:, 0]) # Comparing the graphs, it looks like the 'nearest' interpolation gives the most
# natural looking and smooth results.
```
